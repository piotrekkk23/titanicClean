# 1. Czyszczenie danych

# 1.1 Analiza wizualna zbioru i import danych
Zadanie rozpocznie się od importu danych z pliku tsv. Następnie przy wykorzystaniu funkcji head i tail wyświetlone zostanie pierwsze i ostatnie 15 rekordów. Wyświetlone rekordy pozwolą na wstępną analizę danych znajdujących się w zbiorze.

```r
library(tidyverse)
```

```r
titanic <- read.csv("TitanicMess.tsv", header=T, sep='\t', dec=',', na.strings="")
```

```r
head(titanic,n = 15)
tail(titanic,n = 15)
```

Wykorzystane funkcje dostarczają nam wiele informacji odnośnie zbioru. Widoczne jest że niektóre kolumny są niewłaściwie zaklasyfikowane i brakuje w nich części danych.

# 1.2 Niechciane obserwacje

W zbiorze widoczna jest kolumna ship, określającą statek na którym znajdowali się pasażerowie. Interesują nas tylko osoby znajdujące się na statku Titanic, należy więc sprawdzić czy wszyscy pasażerowie znajdowali się na tym statku. Jeżeli okaże się że wszyscy pasażerowie znajdowali się na jednym statku, kolumnę będzie można usunąć, gdyż nie wpłynie ona na wyniki analizy, ponieważ zawierać będzie tylko jedną wartość. Do sprawdzenia jakie wartości występują w kolumnie wykorzystana zostanie funkcja unique, która zwróci unikalne wartości występujące w kolumnie.

```r
unique(titanic$ship)
```

W kolumnie ship jednyną wartością jest "Titanic" zatem kolumna nie będzie przydatna w analizie i można zostać usunięta. Kolumna zostanie usunięta poprzez wypełnienie kolumny ship wartością NULL.

```r
titanic$ship <- NULL
```

Po usunięciu kolumny Ship zbiór danych zostanie sprawdzony, czy wartości w nim się powtarzają. Wykorzystana zostanie do tego kolumna PassengerId wraz z funkcją duplicated. Funkcja ta wyświetli ID pasażerów które powtarzają się w zbiorze danych.

```r
titanic$PassengerId[duplicated(titanic$PassengerId)]
```

Znaleziono powtarzające się Id 11 oraz 225. Oznacza to że w zbiorze znajdują się 3 rekordy pasażerów z id 11 i 3 rekordy pasażerów z id 225. Rekordy z powtarzającymi się wartościami zostaną usunięte. Do usunięcia wykorzystana zostanie funkcja distinct, która pozwoli na pominięcie pierwszych przepadków występowania danego ID.

```r
titanic <- titanic %>% distinct(PassengerId, .keep_all = TRUE)
```

Kolumna Name określa imię i nazwisko pasażera znajdującego się na statku. Można przyjąć że te informacje nie miały wpływu na szanse na przeżycie katastrofy, kolumna zatem może zostać usunięta.

```r
titanic$Name <- NULL
```

# 1.3 Błędy w strukturze
Zbiór zostanie sprawdzony pod względem występowania w nim błędów w strukturze. Będą one dotyczyły występujących literówek lub wartości z poza określonych kluczy wartości. Kolejny raz zostanie wykorzystana funkcja unique. Pozwoli ona na przeanalizowanie czy wartości w kolumach zgadzają się z opisem zbioru.

```r
unique(titanic$Survived)
unique(titanic$Pclass)
unique(titanic$Sex)
unique(titanic$Embarked)
```

Przeszukiwanie wartości unikalnych dostarczyło wiele przydatnych informacji. Wiemy że kolumny Survived i Pclass zawierają wartości zgodne z kluczem. W kolumnach Sex oraz Embarked pojawiły się literówki. Wartości "malef", "mal" zostaną poprawnione na wartość "male" a wartości "fem" oraz "femmale" zostaną poprawione na wartość "female" w kolumnie Sex, ponieważ ich wygląd świadczy na popełnienie błędu przez osobę tworzącą zbiór danych. Literówki są na tyle podobne do ich prawidłowych wartości kluczy, że możliwa jest ich poprawa. W kolumnie Embarked wartości "So", "Co", "Qe" zostaną poprawione odpowiednio na "S", "C", "Q" by były zgodne z kluczem. Podobnie jak w przypadku kolumny Sex, wartości niezgodne z kluczem są na tyle podobne do ich prawidłowych odpowiedników, że jest możliwa ich poprawa. Do poprawy zostanie wykorzystana funkcja replace, która zamieni błędnie wpisany element występujący na przestrzeni całej kolumny, na ten zgodny z kluczem.

```r
titanic$Sex<-replace(titanic$Sex, titanic$Sex=="malef", "male")
titanic$Sex<-replace(titanic$Sex, titanic$Sex=="mal", "male")
titanic$Sex<-replace(titanic$Sex, titanic$Sex=="fem", "female")
titanic$Sex<-replace(titanic$Sex, titanic$Sex=="femmale", "female")

titanic$Embarked<-replace(titanic$Embarked, titanic$Embarked=="So", "S")
titanic$Embarked<-replace(titanic$Embarked, titanic$Embarked=="Co", "C")
titanic$Embarked<-replace(titanic$Embarked, titanic$Embarked=="Qe", "Q")
```

Wcześniejsza analiza wskazywała, że niektóre wartości kolumn zostały źle zaklasyfikowane.

```r
head(titanic,n = 1)
```

Kolumny Age oraz Fare zostały zaklasyfikowane jako pola znakowe. Oznacza to że konieczna jest ich konwersja do wartości numerycznych. Do konwersji wykorzystana zostanie funkcja as.numeric. Klasyfikacja do wartości pól znakowych wynika z złego separatora dzisiętnego. Przecinek zostanie zatem zamieniony na kropkę i kolumny będą klasyfikowane jako wartości liczbowe.

```r
titanic$Age <- as.numeric(sub(",", ".", titanic$Age, fixed = TRUE))
titanic$Fare <- as.numeric(sub(",", ".", titanic$Fare, fixed = TRUE))
```

```r
summary(titanic$Age)
summary(titanic$Fare)
```

Wartości zostały skonwertowane do wartości numerycznych. Konieczne jest natomiast pozbycie się wartości odstających.

# 1.4 Wartości odstające
Poprzednie działania wykazały że dla niektórych rekordów wiek przyjmuje wartość ujemną lub wynosi 4435 lat, oraz koszt niektórych biletów jest ujemny. Konieczne jest usunięcie rekordów zawierających podobne błędy. Wykorzystana zostanie do tego funkcja subset z parametrami, pozwalającymi na zawarcie danych z określonych przedziałów. Dodatkowo wartości z kolumny Fare zostały zaookrąglone do 2 miejsc po przecinku.

```r
titanic <- subset(titanic, titanic$Age>=0 & titanic$Age<=122 | is.na(titanic$Age))
titanic <- subset(titanic, titanic$Fare>=0 | is.na(titanic$Fare))
titanic$Fare <- round(titanic$Fare, 2)
```

```r
summary(titanic$Age)
summary(titanic$Fare)
```

Dla kolumny fare dla pustej wartości zostanie przyjęta mediana wartości, natomiast dla kolumny age wartości puste zostaną zastąpione wartością -1 która nie występuje w zbiorze, dzięki czemu będzie ona widoczna w trakcie analizy jako osobna klasa.

# **2 Analiza danych**
W tej częsci zostanie przeprowadzona analiza eksploracyjna na przygotowanym wcześniej zbiorze danych. Na początek zostanie sprawdzona ilość osób która przeżyła i nie przeżyła katastrofy. Wykorzystana zostanie do tego funkcja sum, pozwalająca zsumować rekordy z określoną wartością.

```r
sum(titanic$Survived == '1')
sum(titanic$Survived == '0')
```

Z dostępnego zbioru danych możemy się dowiedzieć że katastrofę przeżyło 340 osób a zmarło 543 osoby. Istotnym współczynnikiem decydującem na przeżycie może okazać się płeć pasażerów. Sprawdzona zatem zostanie liczba kobiet i mężczyzn która znajdowała się na statku.

```r
sum(titanic$Sex == 'female')
sum(titanic$Sex == 'male')
```

Na statku znajdowało się 312 kobiet i 571 mężczyzn. Ponieważ ilość kobiet i mężczyzn jest różna, określona zostanie procentowa ilość mężczyzn i kobiet którzy przeżyli i nie przeżyli katastrofy.

```r
print ("kobiet przeżyło:")
sum(titanic$Sex == 'female' & titanic$Survived == 1) * 100 / sum(titanic$Sex == 'female')
print ("kobiet nieprzeżyło:")
sum(titanic$Sex == 'female' & titanic$Survived == 0) * 100 / sum(titanic$Sex == 'female')
print ("mężczyzn przeżyło:")
sum(titanic$Sex == 'male' & titanic$Survived == 1) * 100 / sum(titanic$Sex == 'male')
print ("mężczyzn nieprzeżyło:")
sum(titanic$Sex == 'male' & titanic$Survived == 0) * 100 / sum(titanic$Sex == 'male')
```

Uzyskane dane wskazują że kobiety miały dużo większe szanse na przeżycie. Wynika to prawdopodobnie z tego, iż były ewakuwane jako pierwsze. Kolejnym elementem mogącym wpływać na "istotność" pasażerów była klasa w której podróżowali. W zbiorze danych istnieje podział na 3 klasy z czego 1 oznacza najlepszą jakość obsługi i podróżowania a 3 klasa oznacza najgorszą jakość obsługi i podróżowania.

```r
sum(titanic$Pclass == '1')
sum(titanic$Pclass == '2')
sum(titanic$Pclass == '3')
```

Dane wskazują że najwięcej osób znajdowało się w klasie 3 natomiast najmniej w klasie 2. Możliwe jest wykonanie podziału by sprawdzić czy któraś klasa miała pierwszeństwo jeżeli chodzi o udzielenie pomocy i ewakuację.

```r
print ("1 klasa przeżyło:")
sum(titanic$Pclass == 1 & titanic$Survived == 1) * 100 / sum(titanic$Pclass == 1)
print ("2 klasa przeżyło:")
sum(titanic$Pclass == 2 & titanic$Survived == 1) * 100 / sum(titanic$Pclass == 2)
print ("3 klasa przeżyło:")
sum(titanic$Pclass == 3 & titanic$Survived == 1) * 100 / sum(titanic$Pclass == 3)
```

Dane wskazuję że przydział do klas również miał znaczenie przy określeniu szansy na przeżycie w momencie katastrofy. Możliwe jest połączenie 2 wykonanych wcześniej pomiarów i określenie jak płeć i klasa podróżowania wpływały na szanse na przeżycie.

```r
print ("1 klasa kobiet przeżyło:")
sum(titanic$Pclass == 1 & titanic$Survived == 1 & titanic$Sex == "female") * 100 / sum(titanic$Pclass == 1 & titanic$Sex == "female")
print ("1 klasa mężczyzn przeżyło:")
sum(titanic$Pclass == 1 & titanic$Survived == 1 & titanic$Sex == "male") * 100 / sum(titanic$Pclass == 1 & titanic$Sex == "male")
print ("2 klasa kobiet przeżyło:")
sum(titanic$Pclass == 2 & titanic$Survived == 1 & titanic$Sex == "female") * 100 / sum(titanic$Pclass == 2 & titanic$Sex == "female")
print ("2 klasa mężczyzn przeżyło:")
sum(titanic$Pclass == 2 & titanic$Survived == 1 & titanic$Sex == "male") * 100 / sum(titanic$Pclass == 2 & titanic$Sex == "male")
print ("3 klasa kobiet przeżyło:")
sum(titanic$Pclass == 3 & titanic$Survived == 1 & titanic$Sex == "female") * 100 / sum(titanic$Pclass == 3 & titanic$Sex == "female")
print ("3 klasa mężczyzn przeżyło:")
sum(titanic$Pclass == 3 & titanic$Survived == 1 & titanic$Sex == "male") * 100 / sum(titanic$Pclass == 3 & titanic$Sex == "male")
```

Uzyskane wyniki pokazują że kobiety podróżujące pierwszą klasą miały prawie 97% szans na przeżycie, a mężczyźni podróżujący tą samą klasą zaledwie 37%. Zasobność portfela pasażerów miała zatem istotny wpływ na ich szansę na przeżycie. Ostatnim przebadanym aspektem miejącym wpływ na przeżycie będzie wiek pasażerów. Wybrane zostały odpowiednie przedziały wiekowe mogące identyfikować przykładowo dzieci, młodzież oraz ludzi dorosłych.

```r
print ("Wiek od 0 - 12 przeżyło:")
sum(titanic$Age >= 0 & titanic$Age <= 12 & titanic$Survived == 1) * 100 / sum(titanic$Age >= 0 & titanic$Age <= 12)
print ("Wiek od 13 - 24 przeżyło:")
sum(titanic$Age > 12 & titanic$Age <= 24 & titanic$Survived == 1) * 100 / sum(titanic$Age > 12 & titanic$Age <= 24)
print ("Wiek od 25 - 45 przeżyło:")
sum(titanic$Age > 24 & titanic$Age <= 45 & titanic$Survived == 1) * 100 / sum(titanic$Age > 24 & titanic$Age <= 45)
print ("Wiek od 46 - 100 przeżyło:")
sum(titanic$Age > 45 & titanic$Age <= 100 & titanic$Survived == 1) * 100 / sum(titanic$Age > 45 & titanic$Age <= 100)
print ("Wiek nieznany przeżyło:")
sum(titanic$Age == -1 & titanic$Survived == 1) * 100 / sum(titanic$Age == -1)
```

Wiek w przypadku najmłodszych dzieci miał znaczenie w przypadku szansy na ich przeżycie. W pozostałych grupach wiekowaych szansa na przeżycie klasuje się podobnie, zatem możemy stwierdzić iż największy wpływ na szanse na przeżycie katastrofy ma płeć oraz klasa w której podróżowali pasażerowie. Zwiększone szanse na przeżycie miały również najmłodsze dzieci. W najlepszej sytuacji uplasowały się kobiety które podróżowały pierwszą klasą. Aż 97% takich osób przeżyło katastrofę statku Titanic.
